```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 
```

```python
df=pd.read_csv("studentsperformance.csv")
df.head(10)
```

```python
print(df["math score"].mean())
    
```

```python
print(df["math score"].median())
```

```python
print(df["math score"].std())
```

```python
x=df[["math score","reading score","writing score"]].agg(["mean","median","std"])
print (x)
```

```python
course = df["test preparation course"].unique()
test1_results = {}
for subject in ["math score", "reading score", "writing score"]:
        group1 = df[df["test preparation course"] == course[0]][subject]
        group2 = df[df["test preparation course"] == course[1]][subject]
        t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)
        test1_results[subject] = {"T-Statistic": t_stat, "p-value": p_value}
print(test1_results)
```

```python
education_levels = df['parental level of education'].unique()  # المستويات التعليمية المختلفة
anova_results = {}

for subject in ['math score', 'reading score', 'writing score']:
    groups = [df[df['parental level of education'] == level][subject] for level in education_levels]
    f_stat, p_value = stats.f_oneway(*groups)  # تنفيذ اختبار ANOVA
    anova_results[subject] = {'F-statistic': f_stat, 'P-value': p_value}
print(anova_results)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))
subjects = ['math score', 'reading score', 'writing score']

for i, subject in enumerate(subjects):
    plt.subplot(1, 3, i+1)
    sns.barplot(x="parental level of education", y=subject, data=df, palette="coolwarm", ci=None)
    plt.title(f"{subject} by Parental Education Level")
    plt.xticks(rotation=45)  # لتدوير أسماء المستويات التعليمية لتكون أوضح
    plt.xlabel("Parental Education Level")
    plt.ylabel("Average Score")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df.index, df["math score"], label="Math Score", color="blue", linestyle="-", marker="o")
plt.plot(df.index, df["reading score"], label="Reading Score", color="green", linestyle="--", marker="s")
plt.plot(df.index, df["writing score"], label="Writing Score", color="red", linestyle="-.", marker="^")
plt.title("Line Plot for Scores")
plt.xlabel("Student Index")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
subjects = ['math score', 'reading score', 'writing score']
means = df[subjects].mean()
plt.bar(subjects, means, color=["blue", "green", "red"], alpha=0.7)
plt.title("Average Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.show()
```
